### Start with simpler models (and do data validation as part of the same)

Since we have lots of factors(or possible factors) here, we should be careful of the following :

(a) It is easy to build a model with huge dimensionality and hence overfit to training data.

(b) Doing data validation in itself on the entire set of data will be time consuming and possibly not worth the effort(since we will not be using some of the data).

Therefore, we do the following :

(a) Separate training data into both training and test sets and use the former for training/cross validation and the latter for judging out of sample performance.

(b) Do data validation as and when we use data to build out models. This way, we will be doing validation only on necessary columns.

#### Linear regression model using following factors : ['LotArea']

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Imputer
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# NOTE : I am not quite sure why, but it looks like sometimes predictions go -ve (very few). To make sure that
# these cases do not kill the entire model fitting process, I make sure that we pass on the same by computing logarithm
# on absolute value of predicted value.
def evaluate_model_score(my_model, X, Y):
    predictions = my_model.predict(X)
    
    mean_of_squared_error = mean_squared_error(np.log(np.abs(Y)), np.log(np.abs(predictions)))
    return np.sqrt(mean_of_squared_error)
    
```

```python
def evaluate_neg_model_score(my_model, X, Y):
    return (-1) * evaluate_model_score(my_model, X, Y)
    
```

```python

train_data['LotArea'].isnull().sum()
```

```python
X_columns = ['LotArea']
Y_column = 'SalePrice'
```

```python
def cross_val_score_given_model(my_model, X, Y, cv=5):
    cross_val_score1 = cross_val_score(my_model, X, Y, scoring=evaluate_model_score, cv=cv)
    return cross_val_score1.mean()    
```

```python
def build_model_and_cross_validate(train_data, X_columns, Y_column='SalePrice'):
    my_pipeline  = make_pipeline(Imputer(), linear_model.LinearRegression())
    X = train_data[X_columns]
    Y = train_data[[Y_column]]
    my_pipeline.fit(X, Y)


    return cross_val_score_given_model(my_pipeline, X, Y)
```

```python
build_model_and_cross_validate(train_data, X_columns=X_columns)
```

#### Techicalities of lasso.

##### 1. Lasso after standardizing the predictors
We can standardize predictors using StandardScaler from sklearn.preprocessing library and these can be provided as input to the lasso framework from sklearn.

##### 2. Make lasso standardize predictors itself (by setting normalize option as True)
We can make the lasso method standardize predictors by itself by passing normalize option.

Both these methods should ideally yield similar results (if not exact). Let us see how they stock up !


UPDATE : As per https://stackoverflow.com/questions/24098212/what-does-the-option-normalize-true-in-lasso-sklearn-do, we have decided against using the normalize option.

```python
# Generic routine to check lasso (with all standardizations)
def build_lasso_model_and_cross_validate(train_data, 
                                         X_columns=['GrLivArea', 'LotArea'], 
                                         Y_column='SalePrice',
                                         alpha=0.1):

    my_pipe = Pipeline([('scaling', StandardScaler()),
                        ('model', linear_model.Lasso(alpha=alpha))]
                      )
    X = train_data[X_columns]
    Y = train_data[[Y_column]]
    my_pipe.fit(X, Y)
        


    return (my_pipe, cross_val_score_given_model(my_pipe, X, Y))
```

```python
def print_model_stats(pipeline_obj, cross_validation_score):
    print(cross_validation_score)
    clf_model = pipeline_obj.named_steps['model']

    print(clf_model.intercept_)
    print(clf_model.coef_)    
    print(np.sum(clf_model.coef_ != 0))    
    
```

#### Build a pipeline and verify that output matches with the case of no pipeline

Basically, we build the lasso model with and without a pipeline and make sure that the output matches in both cases.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipe = Pipeline([('scaling', StandardScaler()),
                    ('model', linear_model.Lasso(alpha=0.1))]
                  )
X_columns = ['GrLivArea', 'LotArea', 'PoolArea', 'GarageArea']
Y_column = 'SalePrice'

X = train_data[X_columns]
Y = train_data[[Y_column]]
my_pipe.fit(X, Y)
print_model_stats(my_pipe, -1)
predictions = my_pipe.predict(X)
print(predictions[:5])
```

```python
(clf_model, rel_cross_val_score) = build_lasso_model_and_cross_validate(train_data, 
                                                                        X_columns, 
                                                                        Y_column, 
                                                                        alpha=0.1)
#linear_model.Lasso(alpha=0.1))
#print_model_stats(clf_model, rel_cross_val_score)
input_scaler = StandardScaler()
input_scaler.fit(train_data[X_columns])
X = input_scaler.transform(train_data[X_columns])
Y = train_data[[Y_column]]
lin_model = linear_model.Lasso(alpha=0.1)
lin_model.fit(X, Y)

predictions = lin_model.predict(X)
print(predictions[:5])
```
